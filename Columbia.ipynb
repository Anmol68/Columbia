```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.datos.gov.co/widgets/s2ru-bqt6')
time.sleep(5)
```

```python
get_list=[]
```

```python
def gov():
    
# #     WebDriverWait(driver, 60).until(EC.element_to_be_clickable((By.XPATH, "/html/body/form/div[9]/table/tbody/tr[1]/td[3]")))
    department=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[1]/div")
    print(len(department))
    for i in range(len(department)):
        time.sleep(2)
        temp_dict={}
        time.sleep(1)
        department=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[1]/div")
        temp_dict.update({'department':department[i].text})
        city=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[2]/div")
        temp_dict.update({'city':city[i].text})
        provider_code=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[3]/div")
        temp_dict.update({'provider_code':provider_code[i].text})
        hospital_name=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[4]/div")
        temp_dict.update({'hospital_name':hospital_name[i].text})
        provider_type=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[7]/div")
        temp_dict.update({'provider_type':provider_type[i].text})
        time.sleep(1)
        address=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[13]/div")
        temp_dict.update({'address':address[i].text})
        email=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[14]/div")
        temp_dict.update({'email':email[i].text})
        telephone=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[15]/div")
        temp_dict.update({'telephone':telephone[i].text})
        bed_capacity=driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[18]/div")
        temp_dict.update({'bed_capacity':bed_capacity[i].text})
        print(temp_dict)
        time.sleep(1)
        get_list.append(temp_dict)
        time.sleep(1)
```

```python
def loop():
    nextloop = True
    while nextloop:
        department = driver.find_elements_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[1]/div/table/tbody/tr/td[1]/div")
        print(len(department))
        if len(department) == 100:
            gov()
            time.sleep(2)

            next_click = driver.find_element_by_xpath("/html/body/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/div[4]/div[2]/span[1]/button[2]")
            next_click.click()
            time.sleep(4)


        else:
            gov()
            nextloop = False
```

```python
loop()
```

```python
import pandas as pd
# Converting to Dataframe
df = pd.DataFrame(get_list)
df
```
